# Splitting Data

```python
data.columns
```

```python
X = data[ ['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
Y = data[['Yearly Amount Spent']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x1, x2, y1, y2 = train_test_split(X, Y, test_size = 0.2, random_state = 90)

reg = LinearRegression()
reg.fit(x1, y1)

print('intercept: ', *reg.intercept_)
print('coeff: ', *reg.coef_)
```

```python
pred = reg.predict(x1)
```

# Model Prediction

```python
pred_test = reg.predict(x2)

plt.figure(figsize = (10,5))
plt.scatter(pred_test, y2, edgecolor='r')
plt.xlabel('Predicted Value')
plt.ylabel('Y2 Test Case')
plt.title('Model Prediction', pad = 10, )
plt.show()
```

```python
data_coeff = pd.DataFrame(reg.coef_.T, index = X.columns, columns = ["Coefficient"])
data_coeff
```
